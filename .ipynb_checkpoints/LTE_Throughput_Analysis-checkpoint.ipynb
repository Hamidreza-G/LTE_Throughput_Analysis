## LTE Sector Throughput Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import *
import math
from sklearn.cluster import KMeans
import plotly.express as px
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
sector_df = pd.read_excel("Main City Huawei 3.xlsx", usecols = ['Datetime', 'SECTOR', 'UPM', 'DL User Throughput (Mbit/s)', 'Latitude', 'Longitude', 'Azimuth', 'Coverage Type', 'DL SE', 'CQI', 'BW'] )
```

```python
# adding hour as a feature
sector_df['Hour'] = sector_df['Datetime'].dt.hour
# changing hours after midnight to 24 and 25
sector_df['Hour'][sector_df['Hour'] == 0] = 24
sector_df['Hour'][sector_df['Hour'] == 1] = 25

# changing sector coverage type to numerical values
Coverage_Type = {"Capacity" : 1, "Coverage" : 2}
sector_df["Num Coverage Type"] = sector_df["Coverage Type"].map(Coverage_Type)

sector_df.head(3)
```

### Functions for creating sectors

```python
def degree2rad(degrees):
    return degrees * pi / 180


def sec_poly(long, lat, bearing, radius=0.5, vbw=60):
    R = 6378.1  # Radius of the Earth
    rad_bearing = degree2rad(bearing)

    site_lat = math.radians(lat)  # site lat point converted to radians
    site_lon = math.radians(long)  # site long point converted to radians

    coords = []
    n = 5
    t = np.linspace(degree2rad(bearing - (vbw / 2)), degree2rad(bearing + (vbw / 2)), n)
    for brg in t:
        bor_lat = math.asin(math.sin(site_lat) * math.cos(radius / R) + math.cos(site_lat) * math.sin(radius / R) * math.cos(brg))
        bor_lon = site_lon + math.atan2(math.sin(brg) * math.sin(radius / R) * math.cos(site_lat),
                                    math.cos(radius / R) - math.sin(site_lat) * math.sin(bor_lat))

        bor_lat = math.degrees(bor_lat)
        bor_lon = math.degrees(bor_lon)

        coords.append([bor_lon, bor_lat])

    coords.insert(0, [long, lat])
    coords.append([long, lat])

    return (coords)
```

### Creating sector polygons for all sectors in the file in GeoJSON format

```python
radius = 0.2
vbw = 60
GeoJSON_list = []

for i in range(0,len(sector_df)):
    poly_sec_list = sec_poly(sector_df['Longitude'][i], sector_df['Latitude'][i], sector_df['Azimuth'][i], radius, vbw)
    GeoJSON_list.append(dict(type= 'Feature',
                properties= dict(Sector_ID=sector_df['SECTOR'][i],
                                ),
                geometry= dict(type= 'Polygon',
                               coordinates= [poly_sec_list]
                              ),
                Sector_ID= sector_df['SECTOR'][i]
               )            
    )
```

```python
GeoJSON_dict = {'type': 'FeatureCollection', 'features': GeoJSON_list}
```

### Features used for clustering

```python
# making a feature list
features = [
    f for f in sector_df.columns if f not in ('Datetime', 'SECTOR', 'Latitude', 'Longitude', 'Azimuth', 'Coverage Type', 'Site Type', 'DL SE', 'CQI', 'BW' )    
    ]

cluster_df = sector_df[features]
cluster_df.head(3)
```

```python
pd.plotting.scatter_matrix(sector_df[features], figsize=(14,9));
```

```python
# Feature Scalling
from sklearn import preprocessing

Scalling = 'normal' # 'none', 'normal', 'standard'

if Scalling == 'standard':
    scalled_cluster_ar = preprocessing.StandardScaler().fit_transform(cluster_df)
    
elif Scalling == 'normal':
    scalled_cluster_ar = preprocessing.MinMaxScaler().fit_transform(cluster_df)
    
elif Scalling == 'none':
    scalled_cluster_ar = cluster_df

print(scalled_cluster_ar.min())
print(scalled_cluster_ar.max())
```

### Defining the number of clusters using elbow method

```python
# Elbow method for optimal value of k in KMeans
inertia_list = []
for k in np.arange(1, 10):
    kmn = KMeans(n_clusters=k)
    kmn.fit(scalled_cluster_ar)
    inertia_list.append(kmn.inertia_)
    
plt.plot(np.arange(1,10),inertia_list,'ro-')
plt.title('Elbow Method', fontsize=18)
plt.xlabel('Number of clusters', fontsize=13)
plt.ylabel('Inertia', fontsize=13)
plt.show()
```

### Sector clustering

```python
kmn = KMeans(n_clusters=3, random_state=4) 
kmn.fit(scalled_cluster_ar)
```

```python
# # Appending labels to sector_df

sector_df['Label'] = kmn.labels_

Label_Name = {0: "Coverage", 2: "Residential", 1: "Commercial"}

sector_df['Label'] = sector_df['Label'].map(Label_Name)

Color_Label = {"Coverage": "aqua", "Residential": "red", "Commercial": "yellow"}

sector_df['Color'] = sector_df['Label'].map(Color_Label)

sector_df.head(3)
```

### Fitting a 3rd order polinomial to each cluster's data point
#### The intention is to define a threshold for worst sectors of DL User Throughput.
#### Worst sectors would be below the fitted line considering a marginal border.

```python
coverage_df = sector_df[sector_df.Label == "Coverage"]

coverage_df = coverage_df.sort_values(by='UPM')

ys = coverage_df['DL User Throughput (Mbit/s)']
xs = coverage_df['UPM']

plt.ylabel("DL User Throughput (Mbit/s)")
plt.xlabel("UPM")

plt.scatter(xs, ys, s=1) # assign color to each label
plt.show()
```

```python
x = coverage_df.UPM.values
y = coverage_df['DL User Throughput (Mbit/s)'].values
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

#specify degree of 2 for polynomial regression model
#include bias=False means don't force y-intercept to equal zero
poly = PolynomialFeatures(degree=3, include_bias=False)

#reshape data to work properly with sklearn
poly_features = poly.fit_transform(x.reshape(-1, 1))

#fit polynomial regression model
poly_reg_model = LinearRegression()
poly_reg_model.fit(poly_features, y)

#display model coefficients
print(poly_reg_model.intercept_, poly_reg_model.coef_)
```

```python
#use model to make predictions on response variable
y_predicted = poly_reg_model.predict(poly_features)

plt.ylabel("DL User Throughput (Mbit/s)")
plt.xlabel("UPM")

plt.scatter(x, y, s=1) # assign color to each label

#add line to show fitted polynomial regression model
plt.plot(x, y_predicted, color='purple')
```

```python
coverage_df['y_predicted'] = y_predicted
```
